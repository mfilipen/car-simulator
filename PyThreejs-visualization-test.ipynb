```python
import core.pyThreejsVis as plt
import numpy as np
import core.tf as tf
```

```python
state = plt.robotPlt()
state.plotWorld()
```

```python
w = 1
N = 100
T = 4 * np.pi
dt = T/N
t = np.linspace(0, T, num = N)
steerAngle = 0.314231899 * np.ones(np.shape(t))
theta = w*t + np.pi/2

r = 1

x =  r*(np.cos(w*t))
y =  r*(np.sin(w*t))
```

```python
state.showStates(x, y, theta, steerAngle, dt = 0.0)
```
